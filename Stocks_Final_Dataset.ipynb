# Creating the data frame
---



The code below captures historical stock data for the stock tickers above, calculates the daily win-loss value and a binary win-loss indicator, and stores it in the the Sector_Stocks df. I also kept some of the exploratory data that we ended up not including in the final model, which includes all of the individual stock tickers from the S&P 500. We decided against individual stock tickers due to the volitity vs. a particular index.


````python
import pandas as pd
table=pd.read_html('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
df = table[0]
df2 = table[1]
df.info()
keep = df[['Symbol','GICS Sector','Date added']]


sp500 = keep.rename(columns = {'GICS Sector':'Category'})
sp500['Symbol']
stock_tickers = sp500['Symbol'].values.tolist()
stocks = ["SWTSX","NDXT","AAPL","MSFT","GOOG","AMZN","NVDA","TSLA","META","BABA","CRM"]  + stock_tickers
sp_500_df = pd.DataFrame.from_dict(sp500)
sp_500_df['Category'].unique()
#stocks 
````

````python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from plotnine import *
import yfinance as yf
import pandas as pd
import numpy as np
import yfinance as yf
from datetime import datetime

stocks = ["XLE", "XLF", "XLU", "XLI", "XLK", "XLV", "XLY", "IYR", "AAPL", "XLB", "XLP", "SPY", "^DJI", "NDX"]
end_date = end='2023-04-18'

Sector_Stocks = pd.DataFrame()

for stock in stocks:
    data = yf.Ticker(stock).history(start='2022-03-14', end=end_date)
    data['win_loss_value'] = data.Close - data.Open
    data['win_loss'] = np.where(data['win_loss_value'] > 0, 1, 0)
    data['ticker'] = stock
    Sector_Stocks = Sector_Stocks.append(data)

Sector_Stocks.head()
````

````python
# Checking for missing values
#The only nulls that are included are in the Capital Gains column which will be deleted in the consolidated dataset. 

Sector_Stocks.isna().sum()
````

Realized that by running the quick EDA's above that for our modeling purposes that our dataset needed to be transformed.We transformed our dataset from long form to wide form by creating a dummy variable for each ticker. We also removed unnecessary columns to focus on determining the binary win/loss for each day based on the opening and closing prices of each stock.

````python
#creating dummy variables for each ticker and creating a new data frame withonly the needed variables. 
Categories = pd.get_dummies(Sector_Stocks.ticker)
Sector_Stocks=Sector_Stocks.reset_index()

Sector_Stocks.info()
stocks_new = pd.concat([Sector_Stocks[['ticker','win_loss_value','win_loss','Date']], Categories], axis=1)
stocks_new.head()
````

To complete the Stock Data output, we manipulated the data to have only one date field per row, displaying the win/loss (1/0) for each ticker in its own column. By doing this, we were then able to merge the news sentiment data with the stock data.




````python

# create an empty dictionary to hold our wide-form data
data_wide = {}
tickers = set(stocks_new['ticker'])

# loop through each row in the long-form dataset
for index, row in stocks_new.iterrows():
    # check if the Date column already exists in the dictionary
    if row['Date'] not in data_wide:
        # if it doesn't, create a new dictionary with a key for each ticker
        data_wide[row['Date']] = {}
        for ticker in tickers:
            data_wide[row['Date']][ticker] = 0
    # increment the WinLoss value for the corresponding ticker and date
    data_wide[row['Date']][row['ticker']] += row['win_loss']
# convert the dictionary to a wide-form dataframe
stock_wide = pd.DataFrame.from_dict(data_wide, orient='index')

# display the wide-form dataframe
print(stock_wide)
````
